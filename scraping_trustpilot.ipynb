```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time

headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:110.0) Gecko/20100101 Firefox/110.0"}

def scrape_company_reviews(url, total_pages):
    ratings = []
    reviews = []
    dates = []

    for page in range(1, total_pages + 1):
        current_url = f"{url}?page={page}"
        response = requests.get(current_url, headers=headers)
        soup = BeautifulSoup(response.content, "html.parser")

        review_tags = soup.find_all("div", class_="styles_reviewContent__0Q2Tg")
        rating_tags = soup.find_all("div", class_="star-rating_starRating__4rrcf star-rating_medium__iN6Ty")

        for rating_tag, review_tag in zip(rating_tags, review_tags):
            # Extract rating
            rating = rating_tag.img["alt"].split()[1]
            ratings.append(rating)

            # Extract review content
            review_element = review_tag.find("p", class_="typography_body-l__KUYFJ typography_appearance-default__AAY17 typography_color-black__5LYEn")
            review = review_element.text if review_element else ""
            reviews.append(review)

            # Extract date of review
            date_element = review_tag.find("p", class_="typography_body-m__xgxZ_ typography_appearance-default__AAY17")
            date = date_element.text.split(":")[1].strip() if date_element else ""
            dates.append(date)

        # Add a delay between requests
        time.sleep(4)  # Adjust the sleep time (in seconds) as per your preference

        # Print progress after processing each page
        print(f"Processed page {page} of {total_pages}")

    # Create a pandas DataFrame with the scraped data (assuming ratings, reviews, and dates have the same length)
    data = {"Rating": ratings, "Review": reviews, "Date": dates}
    df = pd.DataFrame(data)
    return df

# Scrape data for different companies
companies = [
    ("Currys", "https://uk.trustpilot.com/review/www.currys.co.uk", 6739),
    ("WHSmith", "https://uk.trustpilot.com/review/www.whsmith.co.uk", 1942),
    ("NextPLC", "https://uk.trustpilot.com/review/www.next.co.uk", 4677),
    ("JDSports", "https://uk.trustpilot.com/review/www.jdsports.co.uk", 9352),
    ("HelloFresh", "https://uk.trustpilot.com/review/hellofresh.co.uk", 3376),
    ("WesternUnion", "https://www.trustpilot.com/review/westernunion.com", 1977)
]

for company_name, url, total_pages in companies:
    print(f"Scraping data for {company_name}")
    company_df = scrape_company_reviews(url, total_pages)
    company_df.to_csv(f"{company_name}_reviews.csv", index=False)
```
